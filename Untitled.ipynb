## Torus

Algorithm from http://alis.uniba.sk/storage/dzb/dostupne/FM/2011/2011-FM-18190/18763v1.pdf

```python
n_points = 300

theta = 2*np.pi*np.random.rand(n_points)
phi = np.pi*np.random.rand(n_points)
u = np.random.rand(n_points)

y = 3+np.sin(phi)
y[u<((3-np.sin(phi))/6)]

torus = np.stack((y*np.cos(theta), y*np.sin(theta), np.cos(phi)), axis=1)
torus += np.random.normal(scale=0.1, size=(n_points , 3))
```

## Sphere

Generates aproximately n_points by rejection method 

```python
n_points = 500
n_points_unfiltered = int(n_points*(24/(4*np.pi)))
x = 2*np.random.rand(n_points_unfiltered, 3) - 1
filtered = x[(x**2).sum(axis=1)<1]
sphere = filtered/np.sqrt((filtered**2).sum(axis=1)[:, None])
sphere += np.random.normal(scale=0.1, size=(sphere.shape[0] , 3))
print("Generated {} points.".format(sphere.shape[0]))
```

```python
plt.scatter(sphere[:,0], sphere[:,2])
plt.axis('equal')
plt.show()
```

```python
def plot_bars(dgm, eps):
    """Plot the barcode."""
    eps = eps**2
    inf = float('inf')


    generator = enumerate(dgm)

    for i,p in generator:
        if p.death == inf:
              plt.plot([p.birth, eps], [i,i], color = 'r')
        else:    
              plt.plot([p.birth, p.death], [i,i], color = 'b')
```

```python
import time

start_index = 3
end_index = 30
eps = 0.2
dim = 3

my_time = []
ds_time = []
n_simp = []

for i in range(start_index,end_index):
    print("Loop {}".format(i))
    
    n_points = i*100

    theta = 2*np.pi*np.random.rand(n_points)
    phi = np.pi*np.random.rand(n_points)
    u = np.random.rand(n_points)

    y = 3+np.sin(phi)
    y[u<((3-np.sin(phi))/6)]

    torus = np.stack((y*np.cos(theta), y*np.sin(theta), np.cos(phi)), axis=1)
    torus += np.random.normal(scale=0.1, size=(n_points , 3))
    
    t0 = time.time()
    f = rips_complex(dim, eps, data=torus)
    t1 = time.time()
    n_simp.append(len(f))
    print(len(f))
    my_time.append(t1-t0)
    
    t0 = time.time()
    f = ds.fill_rips(torus, dim, eps)
    t1 = time.time()
    print(len(f))
    ds_time.append(t1-t0)    
```

```python
ratio = [my_time[i]/ds_time[i] for i in range(len(my_time))]
simp_ratio = [my_time[i]/n_simp[i] for i in range(len(my_time))]

plt.plot(range(start_index,end_index), my_time, c='g')
plt.plot(range(start_index,end_index), ds_time, c='r')
plt.show()

plt.plot(range(start_index,end_index), ratio, c='b')
plt.show()

plt.plot(range(start_index,end_index), simp_ratio, c='c')
plt.show()
```

```python
from operator import itemgetter

def intervals_to_weights(intervals, filtration):
    new_int = []
    for interval in intervals:
        i = interval[0]
        j = interval[1]
        if j== float('inf'):
            new_int.append((filtration[i][1]**2, float('inf')))
        else:
            new_int.append((filtration[i][1], filtration[j][1]))
    return sorted(new_int, key=itemgetter(1))
```

```python
def alt_neighbor_graph(dmat, eps):
    graph = dict()
    n = dmat.shape[0]
    for i in range(n):
        graph[i] = set()
    dmat_copy = dmat.copy()
    dmat_copy[dmat_copy > eps] = 0
    indices = dmat_copy.nonzero()
    for i,j in zip(*indices):
        graph[i].add(j)
    return graph
```

## Laplacian Eigenmaps

```python
from rips import _pairwise_dist
```

```python
dist = _pairwise_dist(torus)

eps = 0.5
t = 0.000001
dist[dist>eps] = 0
exp_dist = np.exp(-dist/t)
exp_dist[dist>eps] = 0
D = exp_dist.sum(axis=1)

M = np.identity(dist.shape[0]) - ((1/D)*dist).T
```

```python
eigvalues, eigvectors = np.linalg.eig(M)
```
